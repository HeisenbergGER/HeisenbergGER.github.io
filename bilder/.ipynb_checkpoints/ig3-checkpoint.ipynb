```python
#matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,20,21)
money = 5000
zinseszins = money*(1.05)**x
plt.figure(figsize=(10,10))
plt.plot(x,zinseszins,label="Zinseszins")
normalerzins = money + money*0.05*x
plt.plot(x,normalerzins,label="normaler Zins")

plt.title("Der Zinseszins-Effekt - Startbetrag 5000€ - 5% p.a.")
plt.xlabel("Zeit [Jahre]")
plt.ylabel("Betrag [€]")
plt.legend()
plt.savefig("ig3.png",format="PNG")
```
