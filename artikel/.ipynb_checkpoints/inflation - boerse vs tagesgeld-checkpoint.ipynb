```python
#matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(7,7))

money = 10000
x = np.linspace(0,10,11)

tg = money*(0.985)**x
b = money*(1.03)**x

plt.title("Zinsrendite vs Börsenrendite mit Inflation - Initialbetrag 10.000€")
plt.xlabel("Zeit [Jahre]")
plt.ylabel("Betrag [€]")

plt.plot(x,tg,label="-1.5% p.a.")
plt.plot(x,b,label="3% p.a.")
plt.legend()
plt.savefig("zinsen-vs-boerse.png",format="PNG")
```
